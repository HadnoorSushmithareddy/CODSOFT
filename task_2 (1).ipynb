```python
# Dictionary Operations
my_dict = {'apple': 3, 'banana': 5, 'orange': 2}

# Accessing value by key
print("Number of apples:", my_dict['apple'])

# Adding a new key-value pair
my_dict['grapes'] = 4
print("After adding grapes:", my_dict)

# Removing a key-value pair
del my_dict['banana']
print("After removing banana:", my_dict)

# Iterating over keys and values
print("Keys and their corresponding values:")
for key, value in my_dict.items():
    print(key, "->", value)

# Checking if a key exists
if 'orange' in my_dict:
    print("We have oranges!")

# Getting the length of the dictionary
print("Number of items in the dictionary:", len(my_dict))
print()

# Tuple Operations
my_tuple = (10, 20, 30, 40, 50)

# Accessing elements by index
print("Element at index 2:", my_tuple[2])

# Slicing
print("Slice from index 1 to 3:", my_tuple[1:4])

# Concatenating tuples
new_tuple = my_tuple + (60, 70, 80)
print("Concatenated tuple:", new_tuple)

# Length of the tuple
print("Length of the tuple:", len(new_tuple))

# Count occurrences of a value
print("Count of 30 in the tuple:", new_tuple.count(30))

# Find the index of a value
print("Index of 50 in the tuple:", new_tuple.index(50))
print()

# List Operations
my_list = [5, 10, 15, 20, 25]

# Appending an element
my_list.append(30)
print("After appending 30:", my_list)

# Removing an element by value
my_list.remove(15)
print("After removing 15:", my_list)

# Inserting an element at a specific position
my_list.insert(2, 35)
print("After inserting 35 at index 2:", my_list)

# Pop element from the list
popped_element = my_list.pop()
print("Popped element:", popped_element)
print("List after pop:", my_list)

# Reversing the list
my_list.reverse()
print("Reversed list:", my_list)

# Sorting the list
my_list.sort()
print("Sorted list:", my_list)

# Checking if an element exists in the list
if 20 in my_list:
    print("20 exists in the list")

# Getting the length of the list
print("Length of the list:", len(my_list))


```
